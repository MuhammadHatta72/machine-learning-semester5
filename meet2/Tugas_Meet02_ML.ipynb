```python
import pandas as pd

data = '/content/drive/MyDrive/Kuliah/Semester 5/ML/Pertemuan02/wbc.csv'
df = pd.read_csv(data)
df.head()
```

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df.info()
```

```python
df.isnull().sum()
```

```python
used_variables = [
    "fractal_dimension_se","radius_worst","texture_worst","perimeter_worst",
    "area_worst","smoothness_worst","compactness_worst","concavity_worst",
    "concave points_worst","symmetry_worst" ,"fractal_dimension_worst", "diagnosis"
]

df_used = df[used_variables]
df_used.head()
```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder() # membuat objek dari LabelEncoder
df_used['diagnosis'] = le.fit_transform(df_used['diagnosis'])
df_used.tail(10)
```

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
df_used['fractal_dimension_se'] = std.fit_transform(df_used[['fractal_dimension_se']])
df_used['radius_worst'] = std.fit_transform(df_used[['radius_worst']])
df_used['texture_worst'] = std.fit_transform(df_used[['texture_worst']])
df_used['perimeter_worst'] = std.fit_transform(df_used[['perimeter_worst']])

df_used.head()
```

4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
# import library for splice data
from sklearn.model_selection import train_test_split

# splict data training and testing, with rasio use 8:2
df_train, df_test = train_test_split(df_used, test_size=0.2, random_state=42)

# print size
print("Ukuran Data Training:", df_train.shape)
print("Ukuran Data Testing:", df_test.shape)
```
